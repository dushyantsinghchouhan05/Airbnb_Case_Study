```python
plt.figure(figsize = [14, 6])
sns.set_style('dark')

sns.scatterplot (
    data = df,
    x = 'price',
    y = 'number_of_reviews',
    hue = 'room_type', 
    alpha = 0.5
);

plt.xlabel('Prices of House < Depending upon the varities >')
plt.ylabel('Number of Reviews given to them!')
```

```python
plt.figure(figsize = [14, 6])
sns.boxplot(x = df['neighbourhood_group'], y = df['price']);
```

```python
plt.figure(figsize = [15,5])
sns.histplot(y = df['number_of_reviews'], x = df['price'], binwidth=20);
```

```python
Profit = df.pivot_table (
    values = 'price',
    index = 'host_name',
    aggfunc = sum
)

Profit.sort_values(by='price', ascending=False)
Pro = Profit.head()
Pro
```

```python
sns.barplot(x = df['room_type'], y = df['price'], width = 0.5, estimator = sum, errorbar = None);
```

```python
sns.barplot(x = df['room_type'], y = df['price'], width = 0.5, hue = df['neighbourhood_group'], estimator = sum, errorbar = None);
```

```python
sns.barplot(x = df['room_type'], y = df['price'], width = 0.5, hue = df['neighbourhood_group'], estimator = np.mean, errorbar = None);
```

```python
RT = df['room_type'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(5, 5))
plt.pie(RT.values, labels=RT.index, autopct='%1.1f%%', textprops={'fontsize': 8})
plt.title("Room Type", fontsize=16)

plt.show()
```

```python
NG = df['neighbourhood_group'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(6, 6))
plt.pie(NG.values, labels=NG.index, autopct='%1.1f%%', textprops={'fontsize': 8})
plt.title("Demand as per Location", fontsize=16)

plt.show()
```
